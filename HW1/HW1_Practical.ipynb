```python
from numpy import random
from matplotlib import pyplot as plt
```

# Sample Generation gaussian distribution function

```python
def sample_generation(mu,sigma2,count):
  sigma = sigma2 ** 0.5
  x = random.normal(mu,sigma,count)
  return x
```

# Mu_MLE function

```python
def MLE(x):
  MLE_mu = sum(x)/len(x)
  return MLE_mu   
```

# Mu_MAP function

```python
def MAP(sigma2 , v , beta , x):
  MAP_mu = (sigma2 * v + beta**2 * sum(x))/(len(x)* beta**2 + sigma2) 
  return MAP_mu
```

# Gaussian distribution x ~ N ( mu = 2 , sigma2 = 5 )
## priror mu is gaussian distribution N ( 1 , 2 )

```python
def test (count):
  mle_mu = 0
  map_mu = 0

  for i in range(10000):
    x = sample_generation(2,5,count)

    mle_mu += MLE(x)
    map_mu += MAP(5 , 1 ,2 , x)

  mean_MLE_mu = mle_mu/10000
  mean_MAP_mu = map_mu/10000
  return mean_MLE_mu , mean_MAP_mu
```

# 10 sample from gaussian distribution
# 100 sample from gaussian distribution
# 1000 sample from gaussian distribution




```python
MLE_mu10 , Map_mu10 = test(10)
print ('for 10 sample: MLE_mu =', MLE_mu10 ,', Map_mu =', Map_mu10)
MLE_mu100 , Map_mu100 = test(100)
print ('for 100 sample: MLE_mu =', MLE_mu100 ,', Map_mu =', Map_mu100)
MLE_mu1000 , Map_mu1000 = test(1000)
print ('for 1000 sample: MLE_mu =', MLE_mu1000 ,', Map_mu =', Map_mu1000)
```

# Plot of Mu_MLE , Mu_MAP

```python
plt.scatter([1,2,3],[MLE_mu10,MLE_mu100,MLE_mu1000], color = "purple")
plt.scatter([1,2,3],[Map_mu10,Map_mu100,Map_mu1000], color = "red")
plt.plot([0.9,3.1],[2,2],color = "orange")
```
